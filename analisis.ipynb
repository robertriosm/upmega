```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import xml.etree.ElementTree as ET


def get_df(xml_file):
    # Parsear XML
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Listas para almacenar datos
    veh_ids = []
    waiting_times = []
    durations = []

    # Extraer cada tripinfo
    for trip in root.findall("tripinfo"):
        veh_ids.append(trip.get("id"))
        waiting_times.append(float(trip.get("waitingTime")))
        durations.append(float(trip.get("duration")))

    # Crear DataFrame
    return pd.DataFrame({
        "veh_id": veh_ids,
        "waiting_time": waiting_times,
        "duration": durations
    })



df = get_df("HISTORY/outputsg50p20m8/969.xml")

df.head()
```
