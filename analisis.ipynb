```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pygad
import xml.etree.ElementTree as ET
```

```python
def get_df(xml_file):
    # Parsear XML
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Listas para almacenar datos
    veh_ids = []
    waiting_times = []
    durations = []

    # Extraer cada tripinfo
    for trip in root.findall("tripinfo"):
        veh_ids.append(trip.get("id"))
        waiting_times.append(float(trip.get("waitingTime")))
        durations.append(float(trip.get("duration")))

    # Crear DataFrame
    return pd.DataFrame({
        "veh_id": veh_ids,
        "waiting_time": waiting_times,
        "duration": durations
    })

```

```python
# version 1
model_g50p20m8 = pygad.load("history/g50p20m8")
df_g50p20m8_0 = get_df("history/outputsg50p20m8/0.xml")

df_g50p20m8_0.describe()
```

```python
df_g50p20m8_f = get_df("history/outputsg50p20m8/959.xml")  
df_g50p20m8_f.describe()
```

```python
best_fitness = max(model_g50p20m8.best_solutions_fitness)
best_gen = model_g50p20m8.best_solution_generation

print(f"best gen: {best_gen}, best fitness: {best_fitness}")
```

```python
model_g50p20m8.plot_fitness()
```
